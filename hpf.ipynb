```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("noisy.jpg", cv2.COLOR_BGR2GRAY)
plt.imshow(image)
plt.gray()
plt.show()
kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])
image = cv2.filter2D(image,-1,kernel)
plt.imshow(image)
plt.gray()
plt.show()
cv2.waitKey(0)
```
